```python
def countries(country_name):
    name=data[data['Geography'] == country_name]
    return name.groupby('Product')['Amount'].sum().sort_values(ascending = False).head(5).plot(kind = 'bar',title = country_name)
```

```python
country_list=data['Geography'].unique().tolist()
```

```python
for i in country_list:
    countries(i)
```
